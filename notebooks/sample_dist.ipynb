```python
import pandas as pd
import matplotlib.pyplot as plt
from contextlib import chdir
import seaborn as sns
```

```python
with chdir(".."):
    raw_counts = pd.read_csv("input_data/raw_counts.csv", sep='\t')
```

```python
raw_counts.head()
```

```python
mut_max = max(raw_counts['saved_mutations'])

plt.figure(figsize=(10, 6))

sns.histplot(raw_counts, 
             x='saved_mutations',
             color='skyblue',
             bins=int(mut_max/100),
            )

#plt.hist(mut_counts, bins=300, edgecolor='black', alpha=0.7)
plt.yscale('log')
plt.xlabel("Number of Mutations")
plt.ylabel("Number of Samples")
plt.title("Distribution of total number of SNPs per sample")
plt.xlim(0, 10000)

with chdir('..'):
    plt.savefig('figures/sample_mutation_dist.png', bbox_inches='tight')

plt.show()
```
